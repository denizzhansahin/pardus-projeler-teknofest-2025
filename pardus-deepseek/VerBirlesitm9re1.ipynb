```python
import pandas as pd
import json

def csv_to_alpaca(csv_file_path, output_json_path):
    """
    CSV dosyasını Alpaca formatına dönüştürür ve JSON dosyasına kaydeder.

    Args:
        csv_file_path (str): CSV dosyasının yolu.
        output_json_path (str): Oluşturulacak JSON dosyasının yolu.
    """

    # CSV dosyasını pandas DataFrame'ine oku
    df = pd.read_csv(csv_file_path)

    # Alpaca formatına dönüştür
    alpaca_data = []
    for index, row in df.iterrows():
        alpaca_entry = {
            "instruction": row["Baslik"],  # Başlık bilgisini "instruction" alanına ekle
            "input": row["Ozet"],  # Özet bilgisini "input" alanına ekle
            "output": row["Icerik"]  # İçerik bilgisini "output" alanına ekle
        }
        alpaca_data.append(alpaca_entry)

    # JSON dosyasına kaydet
    with open(output_json_path, "w", encoding="utf-8") as f:
        json.dump(alpaca_data, f, ensure_ascii=False, indent=4)

# Örnek kullanım
csv_file_path = "csv yolu"  # CSV dosyanızın adı
output_json_path = "1alpaca_format.json"  # Oluşturulacak JSON dosyasının adı
csv_to_alpaca(csv_file_path, output_json_path)
```

```python
# prompt: bir CSV içindeki satır sayısını ve sütun bilgilerini öğrenme

import pandas as pd

def csv_info(csv_filepath):
  """
  CSV dosyasının satır sayısını ve sütun bilgilerini yazdırır.

  Args:
    csv_filepath: CSV dosyasının yolu.
  """
  try:
    df = pd.read_csv(csv_filepath)
    print(f"Dosya: {csv_filepath}")
    print(f"Satır Sayısı: {len(df)}")
    print("Sütun Bilgileri:")
    print(df.info())
  except FileNotFoundError:
    print(f"Hata: Dosya bulunamadı - {csv_filepath}")
  except pd.errors.ParserError:
    print(f"Hata: Geçersiz CSV dosyası - {csv_filepath}")


# Örnek kullanım (oluşturulan tüm CSV dosyaları için)
csv_files = [
             "yeni_birlesik_alpaca_dataset.csv"]

for file in csv_files:
  csv_info(file)
```

```python
from datasets import load_dataset
import pandas as pd

ds = load_dataset("habanoz/news-tr-1.8M")

# Datasetin bir bölümünü (örneğin 'train') bir DataFrame'e çevirin
df = pd.DataFrame(ds['train'])  # 'train' kısmını datasetinize göre ayarlayın

# Metni bölme fonksiyonu
def split_text(text):
    mid = len(text) // 2
    return text[:mid], text[mid:]

# Alpaca formatına çevir
alpaca_df = pd.DataFrame({
    "instruction": ["Verilen metni tamamla."] * len(df),
    "input": [split_text(t)[0] for t in df["text"]],
    "output": [split_text(t)[1] for t in df["text"]]
})

# CSV'ye kaydet
alpaca_df.to_csv("2alpaca_format_dataset.csv", index=False, encoding="utf-8")

print("Dönüşüm tamamlandı!")
```
